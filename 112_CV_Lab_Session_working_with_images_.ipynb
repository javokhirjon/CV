# Affine Transformation

```python
import tensorflow as tf
transformation=tf.keras.preprocessing.image.apply_affine_transform(img, theta=45)

plt.imshow(transformation)
plt.show()
```

```python
#zx: Zoom in x diraction
#zy: Zoom in y direction
transformation=tf.keras.preprocessing.image.apply_affine_transform(img, zx=0.15,  zy=.15)

plt.imshow(transformation)
plt.show()
```

```python
#tx=0 With shift,
#ty=0 Height

transformation=tf.keras.preprocessing.image.apply_affine_transform(img, tx=5,  ty=0)

plt.imshow(transformation)
plt.show()
```
